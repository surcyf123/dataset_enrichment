```python
df = pd.read_csv("./results/200-Pygmalion-2-13B-GPTQ.csv")
mean_duration = df['duration'].mean()
print(f'mean_duration {mean_duration}')
total_relevance_pass_rate = (df[df['total_reward'] != 0].shape[0])/len(df)
print(f'relevance_pass_rate {total_relevance_pass_rate}')

#bert
pass_mean_bert = df['bert'].mean()
print(f'bert_raw_avg {pass_mean_bert}')
pass_mean_bert_norm = df['bert_norm'].mean()
print(f'bert_pass_rate {pass_mean_bert_norm}')

#mpnet
pass_mean_dpo = df['mpnet'].mean()
print(f'mpnet_raw_avg {pass_mean_dpo}')
pass_mean_dpo_norm = df['mpnet_norm'].mean()
print(f'mpnet_pass_rate {pass_mean_dpo_norm}')

#dpo
pass_mean_dpo = df['dpo'].mean()
print(f'dpo_mean {pass_mean_dpo}')
pass_mean_dpo_norm = df['dpo_norm'].mean()
print(f'dpo_norm_mean {pass_mean_dpo_norm}')


# rlhf
pass_mean_rlhf = df['rlhf'].mean()
print(f'rlhf_mean {pass_mean_rlhf}')
pass_mean_rlhf_norm = df['rlhf_norm'].mean()
print(f'rlhf_mean_norm {pass_mean_rlhf_norm}')

#reciprocate
pass_mean_reciprocate_reward = df[df['total_reward'] != 0]['reciprocate'].mean()
print(f'reciprocate_reward_mean {pass_mean_reciprocate_reward}')
pass_mean_reciprocate_reward_norm = df[df['total_reward'] != 0]['reciprocate_norm'].mean()
print(f'reciprocate_reward_mean_norm {pass_mean_reciprocate_reward}')
```

```python

# values if PASS
pass_mean_reciprocate_reward = df[df['relevance_filter'] == 1.0]['reciprocate_reward'].mean()
print(f'pass_mean_reciprocate_reward {pass_mean_reciprocate_reward}')
pass_mean_rlhf_reward = df[df['relevance_filter'] == 1.0]['rlhf_reward'].mean()
print(f'pass_mean_rlhf_reward {pass_mean_rlhf_reward}')
pass_mean_combined_reward = df[df['relevance_filter'] == 1.0]['combined_reward'].mean()
print(f'pass_mean_combined_reward {pass_mean_combined_reward}')
failed_mean_reciprocate_reward = df[df['relevance_filter'] == 0.0]['reciprocate_reward'].mean()
print(f'failed_mean_reciprocate_reward {failed_mean_reciprocate_reward}')
failed_mean_rlhf_reward = df[df['relevance_filter'] == 0.0]['rlhf_reward'].mean()
print(f'failed_mean_rlhf_reward {failed_mean_rlhf_reward}')
```
